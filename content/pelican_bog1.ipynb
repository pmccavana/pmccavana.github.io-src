## Setting Up A New Git Repository at St Lukes.

When setting up a remote repository for Git I go to the S: drive first using the GitBash terminal.

One issue to note is that directories that have a space in their name are shown as string. So if they occur in a path that
you are using on the command line, then you should use a string of the path name.

Having moved to the directory where I keep my Git reposotories I create a new one with, git init --bare. Standard practice is
to append a .git to the name of the new remote repository store.

I can then change to the directory on my own PC where I have my local Git repositories. Before making a new repository you need to generate the folder you are going to use. Enter this new directory and convert it to a repository with, git init.


Code or even general files and directories then be added to this repository so version control can begin.

Once content has been added , git add ., will stage all the content ready to be commited. 

The git commit command allows a comment to be added with a -m option, so that the purpose of the commit may be later identified. 

Having stored the commit locally, the remote repository should be updated to reflect the new changes. First though we need to let the local repository know the remote exists, we will give the remote the name "origin". (The local is called by default "master".)

Now the local changes can be sent to the remote repository to reflect the modifications that have occured with the git push command
