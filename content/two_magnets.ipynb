```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

## I did mention two magnets in the last post, but only considered one, so here goes.

```python
# The B field

def Br(x,y,z):
    if abs(x)<=0.025 and abs(y)<=0.025 and abs(z)<=0.0125:
        bx,by,bz = 0,0,0
    else:
        bx,by,bz = 0,0,0
        for i in [0,1]:
            for j in [0,1]:
                for p in [0,1]:
                    s=x-(-1)**i*0.025
                    t=y-(-1)**j*0.025
                    u=z-(-1)**p*0.0125
                    r=np.sqrt(s**2+t**2+u**2)
                    bx += (-1)**(i+j+p)*np.log(r-t)
                    by += (-1)**(i+j+p)*np.log(r-s)
                    bz += (-1)**(i+j+p)*np.arctan(s*t/(u*r))
    return(bx,by,bz)  
```

```python
# I am not sure that this helps, but lets vectorize it anyhow

xyzB = np.vectorize(Br)
```

### This initial setup considers a 20 cm x 20 cm area around the centre of a magnet

```python
ex = np.arange(-0.2,0.201,0.001)
zed = np.arange(-0.2,0.201,0.001)
X,Z = np.meshgrid(ex, zed) # grid of point
Y = np.zeros((len(ex),len(zed)))

magnet = xyzB(X,Y,Z)
```

#### Showing the Z component of the B field

```python
plt.imshow(magnet[2])
```

If the magnets are 6cm apart, the centre point of rotation for a single magnet array will lie 3cm away from magnet surface.
The magnet is surface is 1.25 cm from its centre, so that is 4.25 cm. So from the original symmetric magnet array we need 
to add 8.5 cm so that we can rotate about this centre.

```python

ex = np.arange(-0.2,0.201,0.001)
zed = np.arange(-0.2,0.286,0.001)
X,Z = np.meshgrid(ex, zed) # grid of point
Y = np.zeros((len(zed),len(ex)))

magnet = xyzB(X,Y,Z)
```

magnet contains the x,y,z values of B.
this can be rotated 180 degrees for a magnet of opposite orientation

```python
turn_magnet = [np.rot90(-i,2) for i in magnet]
```

```python
two_magnets = [i + np.rot90(i,2) for i in magnet ]
```

The contours of this plot have just been selected for demonstration, so we get lines spreadout over the image.

```python

cont =[ 0.0005,0.005,0.01,0.05,0.1,0.5,1,3,6,12]
plt.contour(X,Z,(np.sqrt(two_magnets[2]**2+two_magnets[0]**2)),cont)
plt.title('isotesla around permanent magnets')
plt.text(-0.005,0.025,'N');plt.text(-0.005,-0.037,'S');
```

```python
plt.imshow(np.log(np.sqrt(two_magnets[2]**2+two_magnets[0]**2)))
plt.title('magnitude of B around two permanent magnets')
plt.axis('off');
```

## The streamlines are as we would expect.

```python
plt.streamplot(X, Z, two_magnets[0],two_magnets[2])

plt.title('B field lines');
```

### Add now two opposite oriented magnets to see the repulsive behavior

```python
opp_mags = [turn_magnet[i] + magnet[i] for i in np.arange(3)]

plt.streamplot(X, Z, opp_mags[0],opp_mags[2])
```

#### Zoom into the magnets to see the finer details around the magnets.

```python
plt.streamplot(X[180:-180,150:-150], Z[180:-180,150:-150], opp_mags[0][180:-180,150:-150],opp_mags[2][180:-180,150:-150])
```

### Lets see what the force lines look like for the original pair of magnets with the pole.

```python
b_scalar = np.sqrt(two_magnets[2]**2+two_magnets[0]**2)
divb = np.gradient(b_scalar)

plt.streamplot(X, Z, divb[1],divb[0])
```

### We need to zoom in to see more detail around the mathematical poles.

```python
plt.streamplot(X[180:-180,100:-200], Z[180:-180,100:-200], divb[1][180:-180,100:-200],divb[0][180:-180,100:-200])
```

#### It looks like there is another area 9cm away where opposite force lines meet. These opposite force lines come from our mathematical pole.
